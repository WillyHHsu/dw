```python
import pandas as pd

from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker

from sqlalchemy import MetaData,Column, String, Integer, Date, Table, ForeignKey, Float, VARCHAR,TIMESTAMP,create_engine
```

```python
engine = create_engine('postgresql://pagarme:pagarme@localhost:5432/db', echo =True)
session = sessionmaker(bind=engine)
connection = engine.connect()
session=session()
meta = MetaData()
```

### create tables

```python
transactions = Table(
   'transactions', meta, 
    Column('transaction_id', VARCHAR, primary_key = True), 
    Column('transaction_created_at', TIMESTAMP), 
    Column('transaction_updated_at', TIMESTAMP),
    Column('acquirer_name', VARCHAR),
    Column('company_id', VARCHAR),
    Column('transaction_installments', Integer),
    Column('amount', Integer),
    Column('payment_method', VARCHAR),
    Column('gateway_fee', Float),
    Column('status', VARCHAR),
)

payables = Table(
   'payables', meta, 
    Column('payable_id', VARCHAR, primary_key = True), 
    Column('transaction_id', VARCHAR), 
    Column('payabtransactions_df.sample(1)le_created_at', TIMESTAMP),
    Column('company_id', VARCHAR),
    Column('amount', Integer),
    Column('fee', Integer),
    Column('installment', Integer),
    Column('type', VARCHAR),
    Column('recipient_id', VARCHAR),
    Column('payment_method', VARCHAR),
)

companies = Table(
   'companies', meta, 
    Column('company_id', VARCHAR, primary_key = True), 
    Column('company_created_at', TIMESTAMP), 
    Column('company_mcc', VARCHAR),
    Column('company_type', VARCHAR),

)

transaction_operations = Table(
   'transactionOperations', meta, 
    Column('transaction_operation_id', VARCHAR, primary_key = True), 
    Column('transaction_id', VARCHAR), 
    Column('status', VARCHAR),
    Column('type', VARCHAR),
    Column('acquirer_name', VARCHAR),
    Column('transaction_operation_created_at', TIMESTAMP)
)

meta.create_all(engine)
```

### Transactions

```python
transactions_df = pd.read_csv('raw/Transactions.csv')
for i in transactions_df.iterrows():
    
    _ = transactions.insert().values(transaction_id = i[1]['transaction_id']
                            ,transaction_created_at = i[1]['transaction_created_at']
                            ,transaction_updated_at = i[1]['transaction_updated_at']
                            ,acquirer_name = i[1]['acquirer_name']
                            ,company_id = i[1]['company_id']
                            ,transaction_installments = i[1]['transaction_installments']
                            ,amount = i[1]['amount']
                            ,payment_method = i[1]['payment_method']
                            ,gateway_fee = i[1]['gateway_fee']
                            ,status = i[1]['status']
                            )
    connection.execute(_)
```
